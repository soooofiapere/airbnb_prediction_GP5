# Airbnb (Амстердам 2025)

## 0. Описание выбора компании и формулировка двух связанных бизнес-задач

Сервис Airbnb предоставляет возможность владельцам жилья размещать объявления о краткосрочной аренде недвижимости. Для людей, которые сдают жилье важным является понимание факторов, влияющих на коммерческую успешность объекта: уровня цены, характеристик жилья, расположения, удобств и пользовательских отзывов.

В рамках проекта рассматриваются данные по объявлениям Airbnb в Амстердаме, опубликованные в открытом наборе данных [Inside Airbnb](https://insideairbnb.com/get-the-data/#:~:text=Amsterdam%2C%20North%20Holland%2C%20The%20Netherlands).

Основная цель проекта заключается в прогнозировании коммерческой успешности объявлений Airbnb. При этом в качестве критерия успешности используется не абсолютная годовая выручка, поскольку высокий доход часто обусловлен исключительно высокой стоимостью аренды. Вместо этого рассматривается показатель относительной эффективности, отражающий способность объявления генерировать доход относительно своей цены за ночь.

### Бизнес-задача 1. Оценка успешности объявления по его характеристикам

При создании нового объявления владелец жилья должен определить цену, заполнить описание объекта, указать доступные удобства и условия бронирования. Однако на этом этапе отсутствует информация о будущих бронированиях и фактической доходности объекта.

Разрабатываемая модель позволяет прогнозировать коммерческую успешность объявления на основе его структурированных характеристик: расположения, типа жилья, вместимости, стоимости проживания, удобств и других параметров. Такой прогноз может использоваться до публикации объявления или при его обновлении для оценки потенциальной эффективности объекта.

В рамках бизнес-процесса модель может выступать инструментом поддержки принятия решений. Полученный прогноз помогает владельцу определить, насколько конкурентоспособным является объявление относительно аналогичных объектов, а также выявить характеристики, которые могут быть улучшены для повышения коммерческой эффективности.

В рамках данной задачи модели необходимо предсказать значение целевой переменной `high_revenue_efficiency` по структурированным характеристикам объявления. В качестве признаков используются район размещения, тип жилья, вместимость, цена, набор удобств, а также ограничения и условия бронирования. Для решения задачи будет применяется полносвязная нейронная сеть (Multilayer Perceptron, MLP).

### Бизнес-задача 2. Оценка успешности объявления по отзывам гостей

Отзывы гостей содержат информацию о реальном опыте проживания и отражают аспекты, которые не всегда представлены в структурированных характеристиках объявления. К таким аспектам относятся качество коммуникации с хозяином, чистота жилья, удобство расположения, соответствие ожиданиям гостей и другие факторы, влияющие на удовлетворённость клиентов.

Разрабатываемая модель позволяет прогнозировать коммерческую успешность объявления на основе текстов отзывов. Это даёт возможность определить, насколько пользовательский опыт связан с эффективностью объекта и какие темы чаще встречаются в отзывах успешных объявлений.

В бизнес-процессе такое решение может использоваться для анализа обратной связи от гостей и выявления направлений для улучшения качества сервиса. Результаты модели помогают владельцам понять, какие аспекты проживания оказывают наибольшее влияние на успешность объекта и требуют особого внимания.

В данной задаче модели будет предсказывать ту же целевую переменную `high_revenue_efficiency` на основе текстов отзывов гостей.

В дальнейшем исследование может быть расширено за счет включения данных из других городов и стран. Это позволит построить более универсальную модель, способную выявлять закономерности рынка краткосрочной аренды независимо от конкретной локации. Также увеличение объёма данных может повысить обобщающую способность модели и устойчивость результатов. Однако расширение исследования на множество городов и стран приведет к существенному увеличению объема данных, что потребует больших вычислительных ресурсов, более длительного обучения моделей и дополнительных затрат на хранение и обработку информации.

### 2.2. Предобработка `Listings`

В Listings есть много служебных колонок: ссылки, id и пустые поля. Для модели они не несут большого смысла, поэтому дальше оставим только признаки, которые описывают жилье, владельца жилья, район, цену и текст объявления

```python
listings.sample(5, random_state=42)
```

```python
url_cols = [col for col in listings.columns if "url" in col]

service_cols = [
    "scrape_id",
    "last_scraped",
    "source",
    "calendar_updated",
    "calendar_last_scraped",
    "host_thumbnail_url",
    "host_picture_url",
    "picture_url",
]

drop_now = url_cols + service_cols
drop_now
```

```python
listings = listings.drop(columns=drop_now)
```

```python
listings[["price", "host_response_rate", "host_acceptance_rate"]].sample(5, random_state=42)
```

Некоторые числовые признаки хранятся в строковом формате. В частности, цена записана с символом валюты, а показатели отклика и принятия заявок хоста представлены в процентах. Перед обучением модели приведем их к числовому виду

```python
listings["price"] = (
    listings["price"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .replace("nan", np.nan)
    .astype(float)
)
```

```python
for col in ["host_response_rate", "host_acceptance_rate"]:
    listings[col] = (
        listings[col]
        .astype(str)
        .str.replace("%", "", regex=False)
        .replace("nan", np.nan)
        .astype(float) / 100
    )
```

```python
listings[["price", "host_response_rate", "host_acceptance_rate"]].describe()
```

Пропуски пока не заполняем, сначала соберем все нужные признаки, а уже перед обучением обработаем NaN единообразно

Часть признаков записана как `t` или `f`. Для модели удобнее перевести их в 1/0

```python
bool_cols = [
    "host_is_superhost",
    "host_has_profile_pic",
    "host_identity_verified",
    "instant_bookable",
]
```

```python
for col in bool_cols:
    listings[col] = listings[col].map({"t": 1, "f": 0})
    listings[col] = listings[col].fillna(listings[col].mode()[0])
```

```python
listings[bool_cols].isna().sum()
```

Признак `has_availability` не содержит существенной информации, чтобы различать объявления, поэтому колонку удалим.

```python
listings = listings.drop(columns=["has_availability"])
```

Вместо даты регистрации хоста создадим более интерпретируемый признак – количество лет присутствия на платформе Airbnb

```python
listings["host_since"] = pd.to_datetime(listings["host_since"], errors="coerce")

scrape_date = pd.to_datetime("2025-09-11")

listings["host_years"] = (
    scrape_date - listings["host_since"]
).dt.days / 365
```

```python
listings["host_years"] = listings["host_years"].fillna(
    listings["host_years"].median()
)

listings["host_years"].describe()
```

Поскольку признак `host_years` характеризует стаж владельца жилья на платформе и имеет разброс от недавно зарегистрированных пользователей до владельцев с многолетним опытом, для заполнения пропусков была выбрана медиана как наиболее устойчивая оценка типичного значения признака

Текстовые поля содержат полезную информацию об объекте и его владельце. Для корректной обработки заменим пропущенные значения пустой строкой

Вместо использования полного текста на данном этапе создадим простые количественные характеристики, отражающие объем информации, представленной в объявлении

Для каждого текстового поля рассчитана длина текста, а также признаки наличия описания

Количество и состав удобств являются важными характеристиками объекта. Сформируем признаки, отражающие как общий объем удобств, так и наличие наиболее востребованных сервисов

## 3. Создание новых признаков

### 3.1 Параметры на основе `Calendar`

Поскольку календарь содержит записи по дням, а итоговая модель будет работать на уровне отдельных объявлений, агрегируем данные календаря по каждому объекту размещения. Рассчитаем показатели доступности и ограничения по минимальному и максимальному сроку проживания

```python
calendar_features = (
    calendar
    .groupby("listing_id", as_index=False)
    .agg(
        calendar_days=("date", "count"),
        available_days=("is_available", "sum"),
        unavailable_days=("is_unavailable", "sum"),
        avg_minimum_nights=("minimum_nights", "mean"),
        median_minimum_nights=("minimum_nights", "median"),
        avg_maximum_nights=("maximum_nights", "mean"),
        median_maximum_nights=("maximum_nights", "median"),
    )
)
```

```python
calendar_features
```

Для каждого объявления рассчитали агрегированные характеристики календаря: количество доступных и недоступных дней, а также средние и медианные ограничения на минимальное и максимальное число ночей бронирования

Проверим, отличается ли количество наблюдений календаря между объявлениями

```python
calendar_features["calendar_days"].value_counts()
```

Для всех объявлений календарь содержит одинаковое количество записей – 365 дней. Следовательно, признак `calendar_days` не несет доп информации для модели и может использоваться только как промежуточная величина при расчете долей доступности

Вместо абсолютного количества недоступных дат сформируем более интерпретируемый показатель – долю недоступных дней в календаре объявления. Затем удалим признаки количества доступных и недоступных дней, поскольку содержащаяся в них информация будет отражена в новом признаке

```python
calendar_features["unavailable_rate"] = (
    calendar_features["unavailable_days"] / calendar_features["calendar_days"]
)

calendar_features = calendar_features.drop(
    columns=["calendar_days", "available_days", "unavailable_days"]
)
```

Дополнительно исследуем различия между будними и выходными днями. Для этого отдельно рассчитаем долю недоступных дат в каждой группе дней недели

```python
weekend_stats = (
    calendar
    .groupby(["listing_id", "is_weekend"], as_index=False)
    .agg(
        days=("date", "count"),
        unavailable_days=("is_unavailable", "sum"),
    )
)

weekend_stats["unavailable_rate"] = (
    weekend_stats["unavailable_days"] / weekend_stats["days"]
)

weekend_stats.head()
```

Преобразуем рассчитанные показатели в отдельные признаки, характеризующие недоступность объявления в будние и выходные дни

```python
weekday_features = (
    weekend_stats[weekend_stats["is_weekend"] == 0]
    [["listing_id", "unavailable_rate"]]
    .rename(columns={"unavailable_rate": "weekday_unavailable_rate"})
)

weekend_features = (
    weekend_stats[weekend_stats["is_weekend"] == 1]
    [["listing_id", "unavailable_rate"]]
    .rename(columns={"unavailable_rate": "weekend_unavailable_rate"})
)
```

```python
calendar_features = calendar_features.merge(
    weekday_features,
    on="listing_id",
    how="left"
)

calendar_features = calendar_features.merge(
    weekend_features,
    on="listing_id",
    how="left"
)
```

```python
calendar_features = calendar_features[
    [
        "listing_id",
        "unavailable_rate",
        "weekday_unavailable_rate",
        "weekend_unavailable_rate",
        "avg_minimum_nights",
        "median_minimum_nights",
        "avg_maximum_nights",
        "median_maximum_nights",
    ]
]
```

```python
calendar_features.sample(5, random_state=42)
```

В результате календарные данные были агрегированы до уровня отдельных объявлений. В итоговый набор признаков включены показатели общей, будничной и выходной недоступности, а также характеристики ограничений на минимальную и максимальную продолжительность проживания. Ценовые характеристики не используем, поскольку соответствующая информация уже содержится в датасете Listings и будет получена оттуда при объединении данных

### 3.2. Параметры на основе `Listings`

В качестве целевой переменной сформируем бинарный признак `high_revenue_efficiency`, характеризующий эффективность объявления с точки зрения получаемой выручки относительно стоимости одной ночи проживания.

Для этого рассчитаем показатель:

`revenue_per_price = estimated_revenue_l365d / price`

После этого разделим объявления на две группы по медианному значению данного показателя. Объявления с эффективностью не ниже медианы будут отнесены к классу 1, а остальные – к классу 0.

Поскольку для формирования целевой переменной требуется информация о выручке, наблюдения с отсутствующим значением `estimated_revenue_l365d` исключим из дальнейшего анализа

```python
listings[["price", "estimated_revenue_l365d"]].isna().mean()
```

```python
listings = listings[listings["estimated_revenue_l365d"].notna()].copy()
```

```python
listings["revenue_per_price"] = (
    listings["estimated_revenue_l365d"] / listings["price"]
)

target_median = listings["revenue_per_price"].median()

listings["high_revenue_efficiency"] = (
    listings["revenue_per_price"] >= target_median
).astype(int)

target_median
```

```python
listings["high_revenue_efficiency"].value_counts()
```

```python
listings["high_revenue_efficiency"].value_counts(normalize=True)
```

Полученные классы распределены практически равномерно (51.1% против 48.9%), что позволит избежать существенного дисбаланса классов при обучении модели

Для снижения влияния выбросов и уменьшения асимметрии распределения стоимости проживания дополнительно сформируем логарифмический признак `log_price`

```python
listings["log_price"] = np.log1p(listings["price"])

listings[["price", "log_price", "estimated_revenue_l365d", "revenue_per_price"]].describe()
```

Распределения стоимости и выручки имеют выраженную правостороннюю асимметрию, поэтому логарифмическое преобразование цены позволяет сделать признак более удобным для последующего моделирования

```python
num_cols_to_fill = [
    "host_response_rate",
    "host_acceptance_rate",
    "bedrooms",
    "beds"
]

listings[num_cols_to_fill].isna().mean().sort_values(ascending=False)
```

Наибольшая доля пропусков наблюдается в характеристиках активности хоста (`host_response_rate`, `host_acceptance_rate`), однако она остается умеренной и не требует удаления наблюдений.

```python
for col in num_cols_to_fill:
    listings[col] = listings[col].fillna(listings[col].median())
```

```python
listings["host_response_time"] = listings["host_response_time"].fillna("unknown")
```

Для числовых признаков пропуски были заполнены медианными значениями, что позволяет сохранить объем выборки и снизить влияние возможных выбросов. Для категориального признака `host_response_time` отсутствующие значения выделены в отдельную категорию `unknown`

Исходный признак `property_type` содержит большое количество редких категорий. Для уменьшения размерности пространства признаков объединим категории, встречающиеся менее 30 раз, в общий класс `Other`

```python
property_counts = listings["property_type"].value_counts()

common_property_types = property_counts[property_counts >= 30].index

listings["property_type_grouped"] = listings["property_type"].where(
    listings["property_type"].isin(common_property_types),
    "Other"
)
```

### 3.3. Параметры на основе `Reviews`

```python
reviews.sample(5, random_state=42)
```

На основе даты публикации каждого отзыва были сформируем дополнительные временные признаки

```python
scrape_date = pd.to_datetime("2025-09-11")

reviews["review_age_days"] = (
    scrape_date - reviews["date"]
).dt.days

reviews["is_recent_review"] = (
    reviews["date"] >= scrape_date - pd.Timedelta(days=365)
).astype(int)
```

Признак `review_age_days` отражает количество дней, прошедших с момента публикации отзыва до даты выгрузки данных, а бинарный признак `is_recent_review` показывает, был ли отзыв оставлен в течение последнего года. Эти характеристики позволят учитывать актуальность обратной связи пользователей.

Поскольку модель строится на уровне отдельных объявлений, данные об отзывах необходимо агрегировать от уровня одного отзыва к уровню объявления. Для каждого объекта размещения рассчитыаем показатели, характеризующие количество отзывов, их объем и временную динамику

```python
review_features = (
    reviews
    .groupby("listing_id", as_index=False)
    .agg(
        total_reviews_from_file=("id", "count"),
        avg_comment_len=("comment_len", "mean"),
        median_comment_len=("comment_len", "median"),
        avg_comment_word_count=("comment_word_count", "mean"),
        median_comment_word_count=("comment_word_count", "median"),
        first_review_date=("date", "min"),
        last_review_date=("date", "max"),
        avg_review_age_days=("review_age_days", "mean"),
        recent_reviews_count=("is_recent_review", "sum"),
    )
)
```

Дополнительно вычислим период активности отзывов – интервал между первым и последним отзывом для каждого объявления. Данный показатель позволит оценить длительность присутствия объекта на платформе с точки зрения пользовательской активности

```python
review_features["review_period_days"] = (
    review_features["last_review_date"] - review_features["first_review_date"]
).dt.days
```

Для сопоставления объявлений с различной историей присутствия на платформе рассчитаем среднюю интенсивность получения отзывов в пересчете на один год активности

```python
review_features["reviews_per_year_from_file"] = (
    review_features["total_reviews_from_file"] /
    (review_features["review_period_days"] / 365).replace(0, np.nan)
)
```

```python
review_features["reviews_per_year_from_file"] = (
    review_features["reviews_per_year_from_file"]
    .replace([np.inf, -np.inf], np.nan)
    .fillna(review_features["total_reviews_from_file"])
)
```

```python
review_features[
    [
        "total_reviews_from_file",
        "recent_reviews_count",
        "avg_comment_len",
        "review_period_days",
        "reviews_per_year_from_file",
    ]
].describe()
```

Полученные признаки демонстрируют значительную вариативность между объявлениями. В выборке присутствуют как объекты с единичными отзывами, так и объявления с многолетней историей и большим количеством пользовательских оценок

```python
review_features.sample(5, random_state=42)
```

В результате сформировали таблицу `review_features`, в которой каждой записи соответствует одно объявление Airbnb. Полученные признаки характеризуют объем отзывов, среднюю длину комментариев, интенсивность появления новых отзывов и временную историю взаимодействия пользователей с объектом размещения.

Текстовое содержимое отзывов на данном этапе не используется и будет рассмотрено отдельно в рамках анализа текстовых данных (NLP).

## 5. Сбор финального текстового датасета

Целевая переменная `high_revenue_efficiency`, сформированная ранее, присоединяется к отзывам по идентификатору объявления. Далее из анализа исключим записи без текстового содержимого, поскольку они не несут информации для NLP-моделей.

```python
reviews_with_target = reviews.merge(
    listings_features[["listing_id", "high_revenue_efficiency"]],
    on="listing_id",
    how="inner"
)

reviews_with_target = reviews_with_target[
    reviews_with_target["comments"].str.len() > 0
].copy()
```

Каждому отзыву была сопоставлена целевая переменная соответствующего объявления. Дополнительно были удалены записи с пустыми комментариями.

Поскольку количество отзывов существенно различается между объявлениями, использование каждого отзыва как отдельного объекта наблюдения может привести к смещению выборки в сторону наиболее популярных объектов. Для устранения данной проблемы отзывы агрегируются на уровне объявления

```python
text_data = (
    reviews_with_target
    .sort_values("date")
    .groupby(["listing_id", "high_revenue_efficiency"], as_index=False)
    .agg(
        review_text=("comments", lambda x: " ".join(x.tail(10))),
        review_count=("comments", "count"),
    )
)
```

Для каждого объявления отзывы предварительно отсортировали по дате публикации, после чего объединили в единый текстовый документ. В качестве текстового представления объявления используем десять последних отзывов, чтобы учитывать наиболее актуальный пользовательский опыт. Дополнительно сохраняем количество отзывов, использованных при формировании текста

```python
text_data["high_revenue_efficiency"].value_counts(normalize=True)
```

Распределение целевой переменной после агрегации осталось достаточно сбалансированным: доля успешных объявлений составляет около 57,6%, а доля менее успешных – около 42,4%

```python
text_data[["review_text", "review_count", "high_revenue_efficiency"]].sample(
    5, random_state=42
)
```

Если рассматривать каждый отзыв как отдельный объект наблюдения, объявления с большим количеством отзывов будут многократно представлены в выборке, что приведет к дисбалансу данных и потенциальному смещению модели. Поэтому для NLP-анализа отзывы агрегировали до уровня объявления, а итоговый текст сформировали на основе последних десяти отзывов. Так мы сможем сохранить актуальную информацию о восприятии объекта гостями и одновременно обеспечить сопоставимость наблюдений между объявлениями.
